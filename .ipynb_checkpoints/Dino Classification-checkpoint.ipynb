```python
# calling necessary library
import numpy as np
import matplotlib.pyplot as plt
import os
import cv2

datadir="C:/Users/abhis/Desktop/DS/Dinosaur Classification/Dino Data Set" # setting up the path for the data set
cate=["ankylosaurus","brontosaurus","pterodactyl","trex","triceratops"] # setting up the categories for the image

for category in cate:
    path=os.path.join(datadir,category) # here we are actually declaring the actual path of the image
    for img in os.listdir(path): # here we are iterating through the images, we are converting it to greyscale.
        img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_COLOR)
        plt.imshow(img_array,cmap=None)
        plt.show()
        break
    break
```

```python
print(img_array)
```

```python
# here we can see the images are of different size hence we are going to normalize it./
IMG_SIZE=60

new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
plt.imshow(new_array,cmap='gray')
plt.show()
```

```python
#creating training data set
training_data=[]

def create_training_data():
    for category in cate:
        path=os.path.join(datadir,category) # here we are actually declaring the actual path of the image
        class_num=cate.index(category)
        for img in os.listdir(path):
            try:
                img_array =cv2.imread(os.path.join(path,img),cv2.IMREAD_GRAYSCALE)
                new_array=cv2.resize(img_array,(IMG_SIZE,IMG_SIZE))
                training_data.append([new_array,class_num])
            except Exception as e:
                pass
            
            
create_training_data()

```

```python
print(len(training_data))
```

```python
#shuffling the dataset
import random
random.shuffle(training_data)
```

```python
# printing the sample set

for sample in training_data:
    print(sample[1])
```

```python
x=[]
y=[]
```

```python
for features,label in training_data:
    x.append(features)
    y.append(label)
x=np.array(x).reshape(-1,IMG_SIZE,IMG_SIZE,1)
```

```python
import pickle

pickle_import=open("x.pickle","wb")
pickle.dump(x,pickle_import)
pickle_import.close()

pickle_import=open("y.pickle","wb")
pickle.dump(y,pickle_import)
pickle_import.close()
```

# Creating Neural Network and Classification of Dinos

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense,Dropout,Activation,Conv2D,MaxPool2D,Flatten
import pickle

x=pickle.load(open("x.pickle","rb"))
y=pickle.load(open("y.pickle","rb"))

x=x/255

clf=Sequential()
clf.add(Conv2D(16,3,3,input_shape=x.shape[1:]))
clf.add(Activation("relu"))
clf.add(MaxPool2D(pool_size=(2,2)))
        
clf.add(Conv2D(32,3,3))
clf.add(Activation("relu"))
clf.add(MaxPool2D(pool_size=(2,2)))
        
clf.add(Flatten())
clf.add(Dense(64))

clf.add(Dense(1))
#clf.add(Activation('sigmoid'))

clf.compile(loss="SparseCategoricalCrossentropy",
           optimizer="adam",
           metrics=["accuracy"])

```

```python
X = np.array(x)
y = np.array(y)
clf.fit(X, y, batch_size=8,epochs=3, validation_split=0.1)
```
